```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
def softmax(x, temperature=1.0):
    """Compute softmax with temperature scaling."""
    x = np.array(x)
    exp_x = np.exp((x - np.max(x)) / temperature)  # Subtract max for numerical stability
    return exp_x / exp_x.sum(axis=0, keepdims=True)
```

```python
# Example usage
logits = np.array([2.0, 1.0, 0.1])

# High temperature (softens distribution)
softmax_high_temp = softmax(logits, temperature=5.0)
print("High Temperature Softmax:", softmax_high_temp)

# Low temperature (sharpens distribution)
softmax_low_temp = softmax(logits, temperature=0.5)
print("Low Temperature Softmax:", softmax_low_temp)

# Default temperature (T=1)
softmax_default = softmax(logits)
print("Default Softmax:", softmax_default)
```

```python
plt.plot(range(len(logits)), softmax_default, label = "temp = 0")
plt.plot(range(len(logits)), softmax_low_temp, label = "temp = 0.5")
plt.plot(range(len(logits)), softmax_high_temp, label = 'temp = 2.0')
plt.legend()
plt.show()
```

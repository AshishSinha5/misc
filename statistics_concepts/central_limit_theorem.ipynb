```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
population_size = 1000000
population = np.random.rand(1000000) # uniform distribution [0,1)
```

```python
number_of_samples = 10000
sample_means = np.zeros(number_of_samples)
sample_size = 1
```

```python
for i in range(number_of_samples):
 c = np.random.randint(1,population_size,sample_size)
 sample_means[i] = population[c].mean()
```

```python
def plot_sample_mean(sample_size, sample_means):
    plt.suptitle(f'Sample Size = {sample_size}')
    plt.subplot(1,2,1)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    sns.distplot(sample_means,bins=int(180/5),hist = True,kde = False)
    plt.title('Histogram of Sample mean',fontsize=20)
    plt.xlabel('Sample mean',fontsize=20)
    plt.ylabel('Count',fontsize=20)
    plt.subplot(1,2,2)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    sns.distplot(sample_means,hist = False,kde = True)
    plt.title('Density of Sample mean',fontsize=20)
    plt.xlabel('Sample mean',fontsize=20)
    plt.ylabel('Density',fontsize=20)
    plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
    plt.show()
```

```python
sample_size = range(1, 20, 2)
for size in sample_size:
    sample_means = np.zeros(number_of_samples)
    for i in range(number_of_samples):
        c = np.random.randint(1,population_size,size)
        sample_means[i] = population[c].mean()
    plot_sample_mean(size, sample_means)
```

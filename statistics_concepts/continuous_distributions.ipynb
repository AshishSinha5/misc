```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

### Normal Distribution

```python
x = np.linspace(1, 4, 1000)
y = stats.norm.pdf(x, loc=2, scale=0.1)
sample = np.round(stats.norm.rvs(loc=3.1, scale=0.1, size=10), 3)
print(np.mean(sample))
# plot y vs x in seaborn
sns.lineplot(x = x, y = y)
# x label, y label, title
plt.xlabel('x')
plt.ylabel('pdf')

plt.vlines(x=np.mean(sample), ymin=0, ymax=4, color='red', linestyles='--', label='sample mean')
plt.legend()
plt.title('Normal Distribution with loc=2, scale=0.1')
plt.show()
```

```python
1- stats.norm.cdf(np.mean(sample), loc=2, scale=0.1)
```

## Beta Distribution

```python
def plot_beta_dist(alpha, beta):
    x = np.linspace(0, 1, 100)
    y = stats.beta.pdf(x, alpha, beta)
    # plot the probability mass  using seaborn
    ax = sns.lineplot(x = x, y = y)
    ax.set(xlabel='x', ylabel='pdf(x)', title=f'Beta Distribution {alpha = }, {beta =}')
    plt.show()
```

```python
range(1, 5, 10)
```

```python
alpha_list  = np.arange(0.1, 5, 0.5)
beta_list = np.arange(0.1, 5, 0.5)
# pdf of beta distribution

for alpha in alpha_list:
    for beta in beta_list:
        plot_beta_dist(alpha, beta)

```

```python
# create pyplot figure with 10 * 10 grid
fig, ax = plt.subplots(10, 10, figsize=(30, 30), sharey = True, tight_layout = True)
# iterate over the grid
for i, alpha in enumerate(alpha_list):
    for j, beta in enumerate(beta_list):
        # create a beta distribution
        x = np.linspace(0, 1, 100)
        y = stats.beta.pdf(x, alpha, beta)
        # plot the probability mass  using seaborn
        ax[i, j].plot(x, y)
        # add margins k
        ax[i, j].set_title(f'{alpha = }, {beta =}')

        
```

```python
alpha, beta = 1, 1
x = np.linspace(0, 1, 100)
y = stats.beta.pdf(x, alpha, beta)
# plot the probability mass  using seaborn
plt.plot(x, y)
# add margins k
plt.title(f'{alpha = }, {beta =}\n All the probabilities are equally likely (unconfident probability)')
```

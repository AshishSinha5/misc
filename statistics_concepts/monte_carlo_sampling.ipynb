```python
import numpy as np
import matplotlib.pyplot as plt
```

Monte Carlo methods are a class of techniques for randomly sampling a probability distribution.
Uses - 

    - Estimate Density - gather samples to approximate the distribution of a target distribution 
    - approximating a density
    - optimize a function - isolate the sample that maximizes or minimizes the target function 

E.g. - Draw a sample of a given size and estimate the density

```python
mu = 50
sigma = 5

sizes = [10, 50, 100, 200, 500, 1000]
samples = []
for i, size in enumerate(sizes):
    sample = np.random.normal(mu, sigma, size = size)
    samples.append(sample)
    plt.subplot(3,2,i+1)
    plt.hist(sample)
    plt.title(f'# Samples = {size}')
    plt.xticks([])
plt.show()
```

## Normality Tests

```python
# QQ Plots
from statsmodels.graphics.gofplots import qqplot
```

```python
for i, sample in enumerate(samples):
    qqplot(sample, line = 's')
    plt.title(f'# Samples = {sizes[i]}')
```

```python
## Shapiro-Wilk Test
from scipy.stats import shapiro
```

$H_0 :$ The distribution follows normal distrituion

$H_1 :$ The distribution does not follow the normal distribution 

$\alpha = 0.05$

```python
alpha = 0.05
for i, sample in enumerate(samples):
    stat, p = shapiro(sample)
    if p > alpha:
        print('Fail to reject the null hypothesis - the sample follows the normal distribution')
    else:
        print('Reject the Null hypothesis - the sample does not follow the normal distribution')
```

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# define kernel
def linear_kernel(x, y):
    assert x.shape == y.shape
    return np.dot(x, y)

def squared_exponential_kernel(x, y, alpha = 0.5):
    assert x.shape == y.shape
    assert alpha > 0
    return np.exp(-alpha * np.linalg.norm(x - y) ** 2)

def ornstein_uhlenbeck_kernel(x, y, alpha = 0.5):
    assert x.shape == y.shape
    assert alpha > 0
    return np.exp(-alpha * np.linalg.norm(x - y))

def periodic_kernel(x, y, alpha = 0.5, beta = 0.5):
    assert x.shape == y.shape
    assert alpha > 0
    assert beta > 0
    return np.exp(-alpha * np.linalg.norm(x - y) ** 2) * np.cos(beta * np.linalg.norm(x - y))

def symmetric_kernel(x, y, alpha = 0.5):
    assert x.shape == y.shape
    assert alpha > 0
    return np.exp(alpha * np.min(np.abs(x - y), np.abs(x + y)) ** 2)



```

```python
# chose the points to sample from the function
points = np.arange(0, 1, 0.05).T
U, V = np.meshgrid(points, points)
x = np.vstack([U.flatten(), V.flatten()]).T
n = x.shape[0]
print(f'{points = }\n{U.shape = }\n{V.shape = }\n{x.shape = }\n{n = }')
```

```python
# chose the functon you want to use
cov_func = linear_kernel
C = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        C[i,j] = cov_func(x[i],x[j])
```

```python
u = np.random.normal(0, 1, n )
A, S, B = np.linalg.svd(C)
print(f'{u.shape = }\n{A.shape = }\n{S.shape = }\n{B.shape = }')
z = A @ np.diag(np.sqrt(S)) @ u
print()
plt.plot(z)
```

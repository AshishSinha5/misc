```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
n = 100 # number of trials
p = 1/6 # probability of success

# plot the probability mass function bar graph using seaborn
sns.barplot(x=np.arange(n+1), y=stats.binom.pmf(np.arange(n+1), n, p))
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.xlim(0, 50)
plt.xticks(np.arange(0, 51, 5))

# round off to 3 decimal places
plt.title(f'Binomial distribution PMF {n = }, {p = :.3f}')
```

```python
# find probability of more than 21 successes
# P(X > 21) = 1 - P(X <= 21)
prob = 1 - stats.binom.cdf(40, n, p)
prob
```

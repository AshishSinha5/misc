Fourier transform (FT) is a transform converts a function into a form that describes the frequencies present in the original function.

Fourier transform takes a time based pattern, measures every possible cycle, and returns the overall "cycle recipe" (amplitude, offset, rotation and speed) for every cycle that was found. 

*Given a smoothie, it finds the recipe* - intuitively how do we manage to do that? We add filters to the smoothie - one for each ingredients.

Fourier Transform does the same - it takes a time based function (smoothie) and filters it as a combination of circular paths (ingredients). 
- Get a time based signal 
- Apply filters to measure each possible circular ingredient.
- Combine to together considering which circular ingredient should be added in what quantity.

Some of the applications of fourier transform are - 
- Used in sound equalizers to separate out the "ingredients" i.e. bass/treble.
- Earthquakes can be decomposed in different frequencies and vibrations and building can be designed to not interact with the higher frequencies.

We are essentially *finding the root causes behind an observed effect.*


## Attributes of a circular path
- How big is the circle - (amplitude/radius)
- How fast do we draw (Frequency)
- Where do we start (phase angle)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_span = np.linspace(0, 20, 1000)
time_span
```

### Deconstructing the constituent frequencies of the signal

```python
freq1 = 4
signal1 = np.sin(2*np.pi*freq1*time_span) + 1 + np.random.random(len(time_span))
plt.title('Signal1')
plt.plot(time_span[:100], signal1[:100])
plt.vlines(x = [1/freq1], ymin=0, ymax=2, linestyles=['dashed'])
```

```python
freq2 = 2
signal2 = np.sin(2*np.pi*freq2*time_span) + 1
plt.title('Signal2')
plt.plot(time_span[:100], signal2[:100])
plt.vlines(x = [1/freq2], ymin=0, ymax=2, linestyles=['dashed'])
```

```python
plt.title('Combined Signal')
comb_signal = signal1 + signal2
plt.plot(time_span, signal1, label = 'signal1')
plt.plot(time_span, signal2, label = 'signal2')
plt.plot(time_span, comb_signal, label = 'comb_signal')
plt.legend()
```

```python
freqfft1 = np.fft.fftfreq(time_span.shape[-1], d = 20/1000)
plt.plot(freqfft1[30:100], np.fft.fft(signal1).real[30:100])
```

```python
plt.plot(freqfft1[30:100], np.fft.fft(signal2).real[30:100])
```

```python
plt.plot(freqfft1[30:100], np.fft.fft(comb_signal).real[30:100])
```
